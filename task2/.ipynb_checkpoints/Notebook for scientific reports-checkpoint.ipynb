* пункт 1
- пункт 2

1. пункт 1

# Формулы

Обычная формула: $x^2 + y^2 = 1$

Выключенная формула:
$$
    x^2 + y^2 = 1
$$

Более сложные конструкции тоже работают:
$$
\begin{cases}
&x^2 = 1,\\
&y^2 = 1,\\
&z^2 = 0.
\end{cases}
$$
