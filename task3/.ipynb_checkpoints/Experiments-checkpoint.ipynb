### 1.1 Зависимость времени работы методов от размерности признакового пространства

#### Будем использовать такие модельные данные

```python
clf = SVM(max_iter=10000)
tmp = clf.generate_data(N=500, diag_cov=[16, 16], mean1=[-5, -5], mean2=[5, 5])
```

#### Параметры

```python
n_tests = 9
n_iterations = 5 # усреднение
n_dimensions = [2 ** x for x in range(4, n_tests+4)]
```

#### Функция, которая делает вклад в график для каждого метода

```python
def auxiliary_function(n_tests=n_tests):
    durations = np.zeros(n_tests)
    for test in range(n_tests):
        dimension = n_dimensions[test]
        universal_array = np.ones(dimension, )
        X, y = clf.generate_data(N=400, diag_cov=16 * universal_array, 
                                 mean1=-5 * universal_array, mean2=5 * universal_array, visualize=False)
        start = time.time()
        for iteration in range(n_iterations):
            f(X, y)
        durations[test] = (time.time() - start) / n_iterations
    
    plt.plot(range(durations.shape[0]), durations, '-o')
```

#### Построение самих графиков

```python
global_start = time.time()
```

```python
plt.figure(figsize=(8, 8))
plt.title('Dependency on data dimension')
plt.xticks(range(len(n_dimensions)), n_dimensions)
plt.ylabel('duration, s')
plt.xlabel('dimension')

f = clf.qp_primal_solver
auxiliary_function(6)
f = clf.qp_dual_solver
auxiliary_function(8)
f = clf.subgradient_solver
auxiliary_function(5)
f = clf.liblinear_solver
auxiliary_function()
f = clf.libsvm_solver
auxiliary_function(8)

plt.legend(['qp_primal_solver', 'qp_dual_solver', 'subgradient_solver', 'liblinear_solver', 'libsvm_solver'], 
           loc=1, bbox_to_anchor=(1.4, 1))
plt.show()
```

```python
print('Build time:', round(time.time() - global_start, 2))
```

### Вывод: полностью написанный студентом метод субградиентного спуска плохо справляется с большими размерностями. Лучше справляется с прямой задачей библиотека cvxopt, однако уже на $2^{10}$ время также начинает быстро расти. Лучше всех справляется метод из LibLinear, что логично, так как он разработан как раз для классификации данных с большим количеством объектов больших размерностей

### 3. Поиск оптимального значения C и gamma

```python
X, y = clf.generate_data(N=800, diag_cov=np.array([[1, 0], 
                                                   [8, 36]]), mean1=[6, 8], mean2=[11, 20])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
C_range = np.logspace(-2, 6, 9)
gamma_range = np.logspace(-6, 2, 9)
def best_C_and_gamma():
    best_C = 0
    best_gamma = 0
    best_score = 0
    results = []
    for C in C_range:
        for gamma in gamma_range:
            clf = SVM(max_iter=1000, C=C, gamma=gamma)
            A = clf.qp_dual_solver(X_train, y_train)
            predict = clf.predict(X_test, A=A)
            res = float(np.sum(predict == y_test)) / X_test.shape[0]
            if res > best_score:
                best_score = res
                best_C = C
                best_gamma = gamma
            results.append(res)
    return best_C, best_gamma, best_score, results
```

```python
clf = SVM(max_iter=10000, C=1.0, gamma=0.1)
A = clf.qp_dual_solver(X_train, y_train)
predict = clf.predict(X_test, A=A)
res = float(np.sum(predict == y_test)) / X_test.shape[0]
```

```python
res
```

```python
global_start = time.time()
```

```python
C, gamma, score, results = best_C_and_gamma()
```

```python
print('Build time:', round(time.time() - global_start, 2))
```

```python
C, gamma, score
```

```python
scores = np.array(results).reshape(len(C_range), -1)
```

```python
plt.figure(figsize=(8, 8))
plt.subplots_adjust(left=.2, right=0.95, bottom=0.15, top=0.95)
plt.imshow(scores, interpolation='nearest', cmap=plt.cm.Blues)
plt.xlabel('gamma')
plt.ylabel('C')
plt.colorbar()
plt.xticks(np.arange(len(gamma_range)), gamma_range, rotation=45)
plt.yticks(np.arange(len(C_range)), C_range)
plt.title('Validation accuracy')
plt.show()
```

### Вывод: с ростом С следует уменьшать gamma и наоборот (точность на диагонали немного преобладает)

### 4. Поиск оптимального шага в методе субградиентного спуска

```python
X, y = clf.generate_data(N=500, diag_cov=[1, 16], mean1=[8, 8], mean2=[11, 16])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### Просто $\alpha$

```python
alpha_list = [10 ** x for x in range(-6, -1)]

plt.figure(figsize=(8, 8))
plt.title('Objective curve')
plt.ylabel('functional')
plt.xlabel('iteration / 20')

for alpha in alpha_list:
    clf = SVM(max_iter=10000, alpha=alpha, use_iteration=False, beta=None)
    b, w = clf.subgradient_solver(X_train, y_train)
    plt.plot(clf.objective_curve[::20], label=str(alpha))

plt.legend(loc='upper right')
plt.yscale('log')
plt.show()
```

```python
alpha_list = [10 ** x for x in range(-6, -1)]

plt.figure(figsize=(8, 8))
plt.title('Accuracy')
plt.ylabel('Score, %')
plt.xlabel('Alpha')
width = 0.3
results = []

for alpha in alpha_list:
    clf = SVM(max_iter=10000, alpha=alpha, use_iteration=False, beta=None)
    b, w = clf.subgradient_solver(X_train, y_train)
    predict = clf.predict(X_test, w, b)
    results.append(float(np.sum(predict == y_test)) / X_test.shape[0] * 100)

plt.bar(range(len(alpha_list)), results, width=width)
plt.xticks(np.arange(len(alpha_list)) + width/2., alpha_list)
plt.show()
```

## Вывод: $10^{-2}$ слишком велик, всё время "проскакиваем" оптимум, $10^{-6}$ и $10^{-5}$ слишком малы, не хватает итераций. Однако видно да и интуитивно кажется, что шаг должен зависеть от номера итерации

### Как видно, такая стратегия крайне чувствительна к подбору параметров, лучшей совокупностью  оказалась пара $\alpha=0.5$  и $\beta=0.9$

### 5. Исследование влияния размера подвыборки в стохастическом субградиенте на скорость сходимости и точность решения

```python
X, y = clf.generate_data(N=500, diag_cov=[1, 16], mean1=[8, 8], mean2=[11, 16])
batch_list = [1, 5, 20, 100, 500]
```

#### Точное решение

```python
clf = SVM(max_iter=1000)
b, w = clf.qp_primal_solver(X, y)
accurate_result = clf.compute_primal_objective(X, y, w, b)
```

```python
global_start = time.time()
```

```python
plt.figure(figsize=(8, 8))
plt.title('Influence of batch size on accuracy')
plt.ylabel('functional minus accurate')
plt.xlabel('iteration / 200')
b, w = 0, 0
for batch in batch_list:
    clf = SVM(max_iter=40000, stochastic=True, batch_size=batch, alpha=0.8, use_iteration=True, beta=None)
    b, w = clf.subgradient_solver(X, y)
    plt.plot(np.array(clf.objective_curve[::200]) - accurate_result, label=str(batch))

plt.legend(loc='upper right')
plt.yscale('log')
plt.show()
```
